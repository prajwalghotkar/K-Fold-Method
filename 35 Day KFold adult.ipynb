```python
# KFold and StratifiedKFold

'''Import Libraries '''

import numpy as np
import pandas as pd
import os
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score,StratifiedKFold,StratifiedShuffleSplit,train_test_split,KFold
```

```python
dataset=pd.read_csv('adult.csv')
display(dataset)
```

```python
# Dataset Details

dataset.info()
```

```python
dataset.describe()
```

```python
# Check for Null Values

dataset.isnull().sum()
```

```python
# Convert Gender Male to 0 and Female to 1

dataset['gender']=dataset['gender'].map({'Male':0,'Female':1})
display(dataset)
```

```python
dataset.dtypes
```

```python
# Covert marital - status 0 and 1

dataset['marital-status']=dataset['marital-status'].replace(['Never-married','Divorced','Separated','Widowed'],0)
dataset['marital-status']=dataset['marital-status'].replace(['Married-AF-spouse','Married-civ-spouse','Married-spouse-absent'],1)
display(dataset)
```

```python
# Convert marital-status to Integer

dataset['marital-status']=dataset['marital-status'].astype(int)
dataset.dtypes
```

```python
dataset
```

```python
# remove columns which are not important

dataset.drop(labels=['workclass','education','occupation','relationship','race','native-country'],axis=1,inplace=True)
display(dataset)
```

```python
# Create Y 

y=dataset['income']
y
```

```python
# Dataset as X

dataset=dataset.drop('income',axis=1)
display(dataset)
```

```python
# Train Test Split

x_train,x_test,y_train,y_test=train_test_split(dataset,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Create the Model

models=[]
models.append(('LR',LogisticRegression()))
models.append(('KNN',KNeighborsClassifier()))
print(models)
```

```python
# Create Logistic Regression in the Normal way

LR=LogisticRegression()
LR.fit(x_train,y_train)
```

```python
# Prediction

y_pred=LR.predict(x_test)
y_pred
```

```python
# Accuracy

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
# Display Unique Values

print(y_test.value_counts())
print(pd.DataFrame(y_pred).value_counts())
```

```python
for name,model in models:
    print(name,model)
```

```python
# Implement K Fold 10 - Fold
result=dict()
for name,model in models:
    kfold=KFold(n_splits=10)
    cv_results=cross_val_score(model,x_train,y_train,cv=kfold,scoring='accuracy')
    result[name]=(cv_results,cv_results.mean(),cv_results.std())
```

```python
# Display Accuracy

for key,value in result.items():
    print('Model Name:',key,'\n Individual Accuracy:\n',value[0],'\nMean:',value[1],'\nSTD:',value[2],'\n')
```

```python
# Implement Stratified KFold 10 - Fold
skresult=dict()
for name,model in models:
    skfold=StratifiedKFold(n_splits=10)
    skcv_results=cross_val_score(model,x_train,y_train,cv=skfold,scoring='accuracy')
    skresult[name]=(skcv_results,skcv_results.mean(),skcv_results.std())
```

```python
# Display Accuracy

for pajju,samu in skresult.items():
    print('Model Name:',pajju,'\n Individual Accuracy:\n',samu[0],'\nMean:',samu[1],'\nSTD:',samu[2],'\n')
```

```python
# Implement Stratified KFold 10 - Fold
sskresult=dict()
for name,model in models:
    sskfold=StratifiedShuffleSplit(n_splits=10)
    sskcv_results=cross_val_score(model,x_train,y_train,cv=sskfold,scoring='accuracy')
    sskresult[name]=(sskcv_results,sskcv_results.mean(),sskcv_results.std())
```

```python
# Display Accuracy

for pajju,samu in sskresult.items():
    print('Model Name:',pajju,'\n Individual Accuracy:\n',samu[0],'\nMean:',samu[1],'\nSTD:',samu[2],'\n')
    
```
